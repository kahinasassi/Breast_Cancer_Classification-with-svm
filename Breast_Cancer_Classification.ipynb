````python
# ON Separe target de dataset pour la données en model
X = df_cancer.drop(['target'], axis = 1)
````

````python
y = df_cancer['target']
````

````python
from sklearn.model_selection import train_test_split
````

````python
# 20% TEST 80% TRAIN
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=41)
````

## 1er modèle SVM

````python
from sklearn.svm import SVC
````

````python
from sklearn.metrics import classification_report, confusion_matrix
````

````python
svc_model = SVC()
````

````python
# ENTRAINEMENT
svc_model.fit(X_train, y_train)
````

````python
# TEST
y_predict = svc_model.predict(X_test)
````

````python
y_predict
````

````python
# EVALUATION DU MODEL
cn = confusion_matrix(y_test, y_predict)
````

````python
 # MATRICE DE CONFUSION
sns.heatmap(cn, annot = True)
````

````python
# NORAMALISATION DES ATTRIBUES
min_train = X_train.min()
range_train = (X_train-min_train).max()
X_train_scaled = (X_train - min_train)/range_train
````

````python
min_test = X_test.min()
range_test = (X_test-min_test).max()
X_test_scaled = (X_test - min_test)/range_test
````

````python
# RETRAIN AVEC LES ATTRIBUE NORMAISE
svc_model.fit(X_train_scaled, y_train)
````

````python
y_predict = svc_model.predict(X_test_scaled)
````

````python
# EVALUATION DU MODELE
cn = confusion_matrix(y_test, y_predict)
sns.heatmap(cn, annot = True)
````

````python
print(classification_report(y_test,y_predict))
````

````python
# implémentation svm kernal rbf
param_grid = {'C': [0.1,1,10,100], 'gamma': [1,0.1,0.01,0.001], 'kernel':['rbf']}
````

````python
from sklearn.model_selection import GridSearchCV
````

````python
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 4)
grid.fit(X_train_scaled, y_train)
````

````python
grid.best_params_
````

````python
grid_prediction = grid.predict(X_test_scaled)
````

````python
cn = confusion_matrix(y_test, grid_prediction)
sns.heatmap(cn, annot = True)
````

````python
print(classification_report(y_test,grid_prediction))
````
